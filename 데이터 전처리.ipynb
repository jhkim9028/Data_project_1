```python
import re
def test(s):
    hangul = re.compile('[^ 가-힣0-9]+') # 한글과 띄어쓰기를 제외한 모든 글자
  # hangul = re.compile('[^ \u3131-\u3163\uac00-\ud7a3]+')  # 위와 동일
    result = hangul.sub('', s) # 한글과 띄어쓰기를 제외한 모든 부분을 제거

    result_2 = hangul.findall(s) # 정규식에 일치되는 부분을 리스트 형태로 저장

    return result
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from selenium import webdriver
from selenium.webdriver import ActionChains
from selenium.webdriver.common.by import By
import time
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
from tqdm.notebook import tqdm

# 한글대응
from matplotlib import rc
%matplotlib inline
plt.rcParams['axes.unicode_minus'] = False
rc('font',family='Malgun Gothic')
```

```python
url = 'https://search.shopping.naver.com/catalog/33797589618?query=%EB%8B%A5%ED%84%B0%EC%98%AC%EA%B0%80%20%EC%83%B4%ED%91%B8&NaPm=ct%3Dlmkn4ck8%7Cci%3D405abed5152aef4490271dc8829d478c6f919693%7Ctr%3Dslsl%7Csn%3D95694%7Chk%3Db5fec763155df726d0e5feb39e3d37e22674dfbd'
driver = webdriver.Chrome('../../driver/116/chromedriver.exe')
driver.get(url)
```

```python
# 점수별 리뷰 가져오기

reviews = []
stars = []

def get_review_star():

    # 1점 ~ 5점까지 돌면서 리뷰 긁어오기
    for j in tqdm(range(4,3,-1)):
        
        # i점 필터 클릭
        # 1점 -> 5점 순서로
        score = driver.find_element(By.CSS_SELECTOR, f'#section_review > div.filter_sort_group__2G1QI > div.filter_evaluation_tap__V1J3G > ul > li:nth-child({j}) > a')
        score.click()
        time.sleep(0.5)
        
        # soup 불러오기
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        
        # i점 총 리뷰수
        
        
        
        
        
        
        total = soup.select('li.filter_on__CqqA4')
        if len(total[0].em.text[1:-1]) > 3:           # cnt = i점 리뷰수
            cnt = int(total[0].em.text[1:-1].replace(',',''))
        else:
            cnt = int(total[0].em.text[1:-1]) 
        x = cnt // 20                                 # x = 총 리뷰수 // 20
        total_page = x + 1                            # total_page = 총 페이지 수
        rest = cnt - (x * 20)                         # rest = 마지막 페이지 리뷰수
        next_set = x // 10                            # next_set = '다음'을 누르는 횟수
        
        
        
        # 10페이지 이하인 경우
        if next_set < 1:
            for k in range(total_page):
                # 페이지 정보 가져오기
                html = driver.page_source
                soup = BeautifulSoup(html, 'html.parser')
        
        
                # 리뷰, 평점
                page_reviews = soup.select('p.reviewItems_text__XrSSf')
                page_stars = soup.select('span.reviewItems_average__0kLWX')
                
                # 마지막 페이지가 아닌 경우
                if len(page_reviews) == 20:
                    for i in range(20):
                        reviews.append(page_reviews[i].get_text())
                        stars.append(page_stars[i].get_text()[-1])
                        
                # 마지막 페이지인 경우
                else:
                    for i in range(len(page_reviews)):
                        reviews.append(page_reviews[i].get_text())
                        stars.append(page_stars[i].get_text()[-1])
                
                # 마지막 페이지인 경우
                if k == total_page-1:
                    pass
                # 마지막 페이지가 아닌 경우
                else:
                    # 다음 페이지
                    next_btn = driver.find_element(By.CSS_SELECTOR, f'#section_review > div.pagination_pagination__JW7zT > a:nth-child({k+2})')
                    next_btn.click()
                    time.sleep(0.5)
        
        # 10페이지 이상인 경우
        else:
            # '다음'을 누르는 횟수+1만큼
            for k in range(next_set+1):
                # 마지막 '다음'을 누른 다음이면
                if k == next_set:
                    rest_page = total_page - (next_set*10)     # 마지막 '다음'누른 후 남은 페이지 수
                    
                    for r in range(3,rest_page+3):
                        
                        # 페이지 정보 가져오기
                        html = driver.page_source
                        soup = BeautifulSoup(html, 'html.parser')
                        
                        # 리뷰와 평점 가져오기
                        page_reviews = soup.select('p.reviewItems_text__XrSSf')
                        page_stars = soup.select('span.reviewItems_average__0kLWX')

                        # 마지막 페이지가 아닌 경우
                        if len(page_reviews) == 20:
                            for i in range(20):
                                reviews.append(page_reviews[i].get_text())
                                stars.append(page_stars[i].get_text()[-1])

                        # 마지막 페이지인 경우
                        else:
                            for i in range(len(page_reviews)):
                                reviews.append(page_reviews[i].get_text())
                                stars.append(page_stars[i].get_text()[-1])

                        time.sleep(0.5)
                        if r != rest_page+2:
                            next_btn = driver.find_element(By.CSS_SELECTOR, f'#section_review > div.pagination_pagination__JW7zT > a:nth-child({r})')
                            next_btn.click()
                            time.sleep(0.5)
                        else:
                            pass

                # 마지막 '다음'을 누르기 전이면   
                else:
                    # 페이지를 넘기면 자동으로 1부터 시작하니까 9바퀴
                    for num in range(2,12):
                        
                        # 페이지 정보 가져오기
                        html = driver.page_source
                        soup = BeautifulSoup(html, 'html.parser')
                        
                        # 리뷰와 평점 가져오기
                        page_reviews = soup.select('p.reviewItems_text__XrSSf')
                        page_stars = soup.select('span.reviewItems_average__0kLWX')

                        for i in range(20):
                            reviews.append(page_reviews[i].get_text())
                            stars.append(page_stars[i].get_text()[-1])

                        time.sleep(0.5)
                        
                        # 다음 페이지 넘기기
                        # 1 ~ 10페이지인 경우
                        if k< 1:
                            if num != 11:
                                next_btn = driver.find_element(By.CSS_SELECTOR, f'#section_review > div.pagination_pagination__JW7zT > a:nth-child({num})')
                                next_btn.click()
                                time.sleep(0.5)
                            else:
                                next_page = driver.find_element(By.CSS_SELECTOR, '#section_review > div.pagination_pagination__JW7zT > a.pagination_next__3_3ip')
                                next_page.click()
                                time.sleep(0.5)
                        
                        else:
                            if num == 11:
                                next_page = driver.find_element(By.CSS_SELECTOR, '#section_review > div.pagination_pagination__JW7zT > a.pagination_next__3_3ip')
                                next_page.click()
                                time.sleep(0.5)
                            else:
                                next_btn = driver.find_element(By.CSS_SELECTOR, f'#section_review > div.pagination_pagination__JW7zT > a:nth-child({num+1})')
                                next_btn.click()
                                time.sleep(0.5)
    return [reviews, stars]
```
